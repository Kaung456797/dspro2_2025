```python
import sqlite3

path = ''
db_name = 'student_scores.db'

try:
    # DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)
    # SQL（RDBを操作するための言語）を実行するためのカーソルオブジェクトを取得
    cur = conn.cursor()

    sql = '''
    CREATE TABLE students (
        id INT,
        name TEXT,
        subject TEXT,
        score REAL
    );
    '''
    # SQL文の実行
    cur.execute(sql)

except sqlite3.Error as e:
    print("エラーが発生しました:", e)

finally:
    # DBへの接続を閉じる
    conn.close()
```

```python
import sqlite3

path = ''
db_name = 'student_scores.db'

try:
    # DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)
    # SQL（RDBを操作するための言語）を実行するためのカーソルオブジェクトを取得
    cur = conn.cursor()

    sql = "INSERT INTO students (id, name, subject, score) VALUES (?, ?, ?, ?);"
    data = [
        (2, '佐藤', '英語', 92),
        (3, '松本', '理科', 76),
        (4, '山田', '国語', 85)
    ]
# SQL文の実行
    cur.executemany(sql, data)
# 変更をDBに反映させる
    conn.commit()

except sqlite3.Error as e:
    print("エラーが発生しました:", e)

finally:
    # DBへの接続を閉じる
    conn.close()
```

```python
import sqlite3

path = ''
db_name = 'student_scores.db'

try:
    # DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)
    # SQL（RDBを操作するための言語）を実行するためのカーソルオブジェクトを取得
    cur = conn.cursor()

    sql = "SELECT * FROM students;"
    cur.execute(sql)

except sqlite3.Error as e:
    print("エラーが発生しました:", e)

else:
    print("=== 現在の学生データ ===")
    for row in cur:
        # 行データ(row)はタプルなので，アンパックして列データを取得
        id, name, subject, score = row
        print(f"ID:{id}  名前:{name}  科目:{subject}  点数:{score}")

finally:
    # DBへの接続を閉じる
    conn.close()
```

```python
import sqlite3

path = ''
db_name = 'student_scores.db'

try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()

    sql = "UPDATE students SET score = ? WHERE id = ?;"
# SQL文の実行
    cur.execute(sql, (95, 3))
# 変更をDBに反映させる
    conn.commit()

except sqlite3.Error as e:
    print("エラーが発生しました:", e)

finally:
# DBへの接続を閉じる
    conn.close()
```
