```python
x = 1
y =0
```

```python
# ０で割っているのでエラが起きる
x/y
```

```python
try:
    result = x / y
except ZeroDivisionError as e:
    result = None
    print("エラが発生しました："、e)
    print("０で割ることはできません。結果をNoneに設定します。")
    
print("結果：", result)
```
