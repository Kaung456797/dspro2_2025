### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
# 問1：文字列 "datascience" を逆順に並べる

text = "datascience"

# [::-1] を使うと文字列を逆向きに取り出せる
reversed_text = text[::-1]

print("問1の答え：", reversed_text)
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
# 問2：「ヘルシンキエニシア」から 1,3,5,7,9 番目の文字を取り出して連結する
# Pythonは「0から数える」ので、0,2,4,6,8 を取り出す

text2 = "ヘルシンキニシア"

# 指定のインデックスの文字を取り出して連結
indices = [0, 2, 4, 6, 8]
picked = "".join(text2[i] for i in indices)

print("問2の答え：", picked)
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
# 問3：ヘルシンキ + エストニア を交互に連結する（丁寧な説明付き）

a = "ヘルシンキ"
b = "エストニア"

result2 = ""

# 短い方の文字数まで繰り返す
for i in range(min(len(a), len(b))):
    result2 += a[i] + b[i]

# 残りの文字を追加
result2 += a[min(len(a), len(b)):]
result2 += b[min(len(a), len(b)):]

print("問3の答え（丁寧な版）：", result2)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
# 問4
# 文を単語ごとに分解し、各単語をアルファベット順に並べたリストを作る

sentence4 = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# split() を使って単語に分解
words = sentence4.split()

# アルファベット順に並べ替え（sort / sorted）
sorted_words = sorted(words)

print("問4の答え：単語をアルファベット順にしたリスト")
print(sorted_words)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
# 問5
sentence5 = ("It is not the strongest of the species that survives, "
             "nor the most intelligent that survives. "
             "It is the one that is most adaptable to change.")

# 文を単語ごとに分割
words5 = sentence5.replace(",", "").replace(".", "").split()

# 抜き出す単語番号（1番目 → Pythonでは 0番目）
indices = [1, 2, 5, 6, 8, 9, 16]

# 結果を入れる辞書
result_dict = {}

for idx in indices:
    if idx - 1 < len(words5):  # 範囲チェック
        word = words5[idx - 1]         # 1番目 → words5[0] に対応
        head2 = word[:2]               # 先頭2文字
        result_dict[head2] = idx       # 辞書に登録

print("問5の答え：先頭2文字 → 単語の位置 の辞書")
print(result_dict)
```
